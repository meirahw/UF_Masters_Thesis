```python
# make matplotlib plots look like they're part of the notebook
%matplotlib inline

# import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

# import numpy
import numpy as np

# import netcdf
import netCDF4
from netCDF4 import Dataset

# import os
import os
```

```python
# create function that appends lat/lon to ERO netcdf
def append_EROs(ERO_append):

    # print keys of original file
    ERO_append.variables.keys()

    # print dimensions of original file
    ERO_append.dimensions.items()

    # add lat/lon variables with units and name
    lat = ERO_append.createVariable('lat', np.float32, ('lat',))
    lat.units = 'degrees_north'
    lat.long_name = 'latitude'
    lon = ERO_append.createVariable('lon', np.float32, ('lon',))
    lon.units = 'degrees_east'
    lon.long_name = 'longitude'

    # make sure lat/lon exist as variables now; see lat/lon starting points, deltas
    print(ERO_append)

    # create lat and lon variables based on ERO info
    lat[:] = np.linspace(25, 25+0.09*276, 276)
    lon[:] = np.linspace(-129.8, -129.8+0.09*721, 721)

    # make sure lat/lon were added correctly
    print(ERO_append)

    # close file after saving
    ERO_append.close()
```

```python
# append all files with lat and lon in folder
for filename in os.listdir(r'C:\Users\meira\OneDrive - University of Florida\UF_Thesis_Research\ERO_NetCDF\spring_extras'):
    
    ERO_file = Dataset(r'C:\Users\meira\OneDrive - University of Florida\UF_Thesis_Research\ERO_NetCDF\spring_extras\\'+ filename, 'a')
    
    append_EROs(ERO_file)
```
